**2**. Teniendo en cuenta el material teórico y los lenguajes investigados en el punto 1, elija uno de ellos y genere 1000 valores de números aleatorios con distribución U(0,1). Luego realice un histograma para observar gráficamente cómo es la distribución

```python
import numpy as np
import seaborn as sns
```

```python
# 1000 datos aleatorios siguiendo una distribución U(0, 1)
uniform_df = np.random.uniform(low=0.0, high=1.0, size=1000)
sns.distplot(uniform_df)
```

**3**. A partir de la muestra obtenida de 1000 valores en el punto 1, realice:

    a. Una transformación para que la variable aleatoria tenga una distribución de probabilidad Exponencial con parámetro λ = 6.
    
    b. Un gráfico adecuado para variable aleatoria.
    
    c. Genere otros 100 valores aleatorios siguiendo una distribución Exponencial con parámetro λ = 2.
    
    d. Genere 1000 valores aleatorios con las mismas características que en el inciso c.
    
    e. Grafique los incisos c y d.
    
    f. Compare resultados para 100 y 1000 valores describiendo similitudes o diferencias encontradas.

```python
# Transformación de distribución (uniforme -> exponencial con lambda=6)
def _uniform_to_exp(u, lambd):
    return (-1 * np.log(u)) / lambd

uniform_to_exp_df = np.fromiter(
    map(lambda u: _uniform_to_exp(u, 6), uniform_df),
    dtype=np.float
)
sns.distplot(uniform_to_exp_df)
```

```python
# 100 datos aleatorios con distribución exponencial lambda=2
exp_df_1 = np.random.exponential(scale=1/2, size=100)
sns.distplot(exp_df_1)
```

```python
# 1000 datos aleatorios con distribución exponencial lambda=2
exp_df_2 = np.random.exponential(scale=1/2, size=1000)
sns.distplot(exp_df_2)
```

Si bien ambos gráficos presentan una curva descendente, el segundo al tener mayor cantidad de datos, genera un gráfico de distribución con una curva más suave (sin tantos picos como el primero)

**4**. De acuerdo a los resultados obtenidos en el ejercicio anterior (incisos c y d). Especifique:'

    a. ¿Cuál fue la media muestral de los datos generados?

    b. ¿Cuál fue la varianza muestral de los datos generados?

    c. Realice un intervalo de confianza con  tres desvíos estándar (DE) para la media de ambas muestras (incisos c y d), con un nivel de confianza de 95%.

```python
print(f"Media muestral de 100 datos (distribución U(0, 1): {exp_df_1.mean()}")
print(f"Varianza muestral de 100 datos (distribución U(0, 1): {exp_df_1.var()}")
```

```python
print(f"Media muestral de 1000 datos (distribución U(0, 1): {exp_df_2.mean()}")
print(f"Varianza muestral de 1000 datos (distribución U(0, 1): {exp_df_2.var()}")
```

```python
def get_confidence_interval(n, mean, std, z_value=1.96, std_mult=1):
    ci = z_value * ((std_mult * std) / (n ** 0.5))
    return mean+ci, mean-ci
```

```python
lower_bound, upper_bound = get_confidence_interval(100, exp_df_1.mean(), exp_df_1.std())
print(f"Intervalo de confianza para la media de 100 datos aleatorios (siguiendo una distribución exponencial): [{lower_bound}, {upper_bound}]")
```

```python
lower_bound, upper_bound = get_confidence_interval(1000, exp_df_1.mean(), exp_df_1.std())
print(f"Intervalo de confianza para la media de 1000 datos aleatorios (siguiendo una distribución exponencial): [{lower_bound}, {upper_bound}]")
```
